```python
# kafka_producer.py
import json
import time
import pandas as pd
from kafka import KafkaProducer

BOOTSTRAP_SERVERS = ["16.171.35.252:9092"]
TOPIC = "trade_value"          # change if your topic name differs
CSV_PATH = "data/indexProcessed.csv"  # or "indexProcessed.csv"
SLEEP_SEC = 1

producer = KafkaProducer(
    bootstrap_servers=BOOTSTRAP_SERVERS,
    value_serializer=lambda x: json.dumps(x).encode("utf-8"),
)

# quick connectivity test (matches the notebookâ€™s idea of sending a dummy dict)
producer.send(TOPIC, value={"surname": "Phil"})
producer.flush()

df = pd.read_csv(CSV_PATH)

try:
    while True:
        link = df.sample(1).to_dict(orient="records")[0]
        producer.send(TOPIC, value=link)
        producer.flush()
        time.sleep(SLEEP_SEC)
except KeyboardInterrupt:
    pass
finally:
    producer.flush()
    producer.close()
```
